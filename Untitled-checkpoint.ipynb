```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[data.feature_names], df['target'], random_state=666, shuffle=True)
```

```python
print(len(x_train))
print(len(x_test))
```

```python
426/(426+143)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
pred = clf.predict(x_test)
confusion_matrix(y_test,pred)
```

```python
(54 + 77) / (54+2+10+77)
```

```python
print(classification_report(y_test,pred))
```

```python
fn=data.feature_names
cn=data.target_names
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=800)
plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

```python
feature_importance = pd.DataFrame({'Features':df.drop(columns='target').columns,'Importance':clf.feature_importances_})
feature_importance = feature_importance.sort_values(by=['Importance'],ascending=False)
feature_importance
```

```python
feature_importance.head()
```

```python
import seaborn as sns
f, ax = plt.subplots(figsize=(11, 9))
values = np.array(feature_importance.Importance) 
clrs = ['#CCE0FC' if (x < max(values)) else '#A7C9F9' for x in values ]
ax = sns.barplot(x='Features',y='Importance',data=feature_importance.head(),palette=clrs)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
sns.despine()
```
