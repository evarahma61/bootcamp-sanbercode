<h2>Tugas Harian No 2 : Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)

fig = plt.figure()
ax = fig.add_subplot()

plt.show()
```

```python
# Membuat figure dan axis secara explicit

fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, y)
plt.show()

# Membuat figure dan axis secara implicit
plt.plot(x,y)
```

<h2>Tugas Harian No 3 : Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.style.use('seaborn')
plt.plot(x, y, x, y2, x, y3)
```

Expected Output :



```python
fig, ax = plt.subplots()
ax.plot(x, y, x, y2, x, y3)
```
